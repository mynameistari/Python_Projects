```python
#Import necessary libraries and dataset
import pandas as pnd
sales_data= pnd.read_csv("C:/Users/ADMIN/Downloads/q1_dataset.csv")
sales_data.describe()
```

```python
#Display the dataset
sales_data.head(100)
```

```python
#Calculate total sale per entry
sales_data['Sale_value'] = sales_data['price'] * sales_data['quantity']
sales_data.head(100)
```

```python
#Calculate the cummulative sales made for each product
total_sales_per_product= sales_data.groupby('product')['Sale_value'].sum().reset_index()
total_sales_per_product.rename(columns={'Sale_value': 'Total sales'}, inplace= True)
total_sales_per_product.head(6)
```
